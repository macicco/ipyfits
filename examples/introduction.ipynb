Add ability to draw lines

```python
canvas = w._ginga_viewer.add_canvas()
```

```python
canvas.set_drawtype('line', color='yellow', linewidth=5, linestyle='dot', showcap=True)
```

```python
def clear_old_lines(canvas, button, data_x, data_y):
    if canvas.get_draw_mode() != 'edit':
        canvas.delete_all_objects()
    
canvas.set_callback('cursor-down', clear_old_lines)
```

Switch to edit mode; now you can move the line around or rotate/scale it.

```python
canvas.set_draw_mode('edit')
```

Back to draw mode

```python
canvas.set_draw_mode('draw')
```

Get the line, extract the data, and then plot in bqplot.

```python
l1 = canvas.get_objects()[0]
```

```python
img = w._ginga_viewer.get_image()
```

```python
dots = img.get_pixels_on_line(l1.x1, l1.y1, l1.x2, l1.y2)
coordinates = img.get_pixels_on_line(l1.x1, l1.y1, l1.x2, l1.y2, getvalues=False)
```

```python
from bqplot import pyplot as plt
import numpy as np
```

```python
plt.figure(1, title='Line Profile')
plt.plot(dots)
plt.show()
```
